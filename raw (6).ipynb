# 1. OUTLIER DETECTION (Z-SCORE) FOR PM2.5 (µg/m³)

```python
import pandas as pd
from scipy.stats import zscore

# Load your data
df = pd.read_csv('rawdataju.csv')

# Calculate Z-scores for PM2.5
df['PM2.5_Z'] = zscore(df['PM2.5 (µg/m³)'])

# Set a threshold (commonly 3)
threshold = 3

# Identify outliers
outliers = df[(df['PM2.5_Z'].abs() > threshold)]

# Show results
print(f"Number of outliers: {len(outliers)}")
print(outliers[['Timestamp', 'PM2.5 (µg/m³)', 'PM2.5_Z']].head())
print(outliers[['Timestamp', 'PM2.5 (µg/m³)', 'PM2.5_Z']])
```

# 1. OUTLIER DETECTION (IQR) FOR PM2.5 (µg/m³)

```python
import pandas as pd

# Load your data
df = pd.read_csv('rawdataju.csv')

# Column to check
col = 'PM2.5 (µg/m³)'

# Calculate IQR
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Create an IQR flag column
df['PM2.5_IQR_Outlier'] = ((df[col] < lower_bound) | (df[col] > upper_bound))

# Show results
print(f"Number of outliers: {len(outliers)}")
print(outliers[['Timestamp', col]].head())
print(outliers[['Timestamp', col]])
```

# OUTLIER DETECTION FOR SVM

```python
import pandas as pd
import numpy as np
from sklearn.svm import OneClassSVM
import warnings

warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('rawdataju.csv')

# List of columns to process
columns = [
    'PM2.5 (µg/m³)',
    'PM10 (µg/m³)', 
    'NO2 (µg/m³)', 
    'CO (mg/m³)', 
    'Ozone (µg/m³)', 
    'AT (°C)', 
    'RH (%)'
]

# Apply One-Class SVM to each column individually
for col in columns:
    print(f"\n=== Column: {col} ===")
    
    data = df[col].dropna().values.reshape(-1, 1)  # reshape to 2D
    if len(data) < 10:
        print("Not enough data points for SVM.")
        continue

    # Create and fit One-Class SVM model
    svm = OneClassSVM(kernel='rbf', gamma='scale', nu=0.05)
    svm.fit(data)
    
    preds = svm.predict(data)
    # SVM labels: +1 = normal, -1 = outlier
    outliers = data[preds == -1]
    
    print(f"Total values: {len(data)}")
    print(f"Outliers detected: {len(outliers)}")
    print(f"Outlier values: {outliers.flatten().tolist()}")
```

# OUTLIER DETECTION FOR GRUBBS' TEST

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load dataset
df = pd.read_csv('rawdataju.csv')

# List of columns to apply Grubbs' Test
columns = [
    'PM2.5 (µg/m³)',
    'PM10 (µg/m³)', 
    'NO2 (µg/m³)', 
    'CO (mg/m³)', 
    'Ozone (µg/m³)', 
    'AT (°C)', 
    'RH (%)'
]

# Define Grubbs' Test function
def grubbs_test(data, alpha=0.05):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    
    # Grubbs' statistic
    G = max(abs(data - mean)) / std_dev
    
    # Critical value
    t_critical = stats.t.ppf(1 - alpha / (2 * n), n - 2)
    G_critical = ((n - 1) / np.sqrt(n)) * np.sqrt(t_critical**2 / (n - 2 + t_critical**2))
    
    return G, G_critical

# Apply Grubbs' Test to each column
alpha = 0.05  # significance level

for col in columns:
    print(f"\nColumn: {col}")
    data = df[col].dropna().values
    if len(data) < 3:
        print("Not enough data points for Grubbs' Test.")
        continue

    G, G_crit = grubbs_test(data, alpha)
    
    print(f"Grubbs' Statistic (G): {G:.4f}")
    print(f"Critical Value (G_critical): {G_crit:.4f}")

    if G > G_crit:
        # Outlier detected
        outlier_value = data[np.argmax(abs(data - np.mean(data)))]
        print(f"Outlier detected: {outlier_value}")
    else:
        print("No outliers detected.")

```

# OUTLIER DETECTION FOR TIETJEN-MOORE FUNCTION

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load dataset
df = pd.read_csv('rawdataju.csv')

# List of columns to process
columns = [
    'PM2.5 (µg/m³)',
    'PM10 (µg/m³)', 
    'NO2 (µg/m³)', 
    'CO (mg/m³)', 
    'Ozone (µg/m³)', 
    'AT (°C)', 
    'RH (%)'
]

# Define the Tietjen-Moore function
def tietjen_moore_test(data, k, tail='largest', alpha=0.05):
    n = len(data)
    data_sorted = np.sort(data)
    mean_full = np.mean(data_sorted)
    sq_dev_full = np.sum((data_sorted - mean_full) ** 2)

    if tail == 'largest':
        reduced_data = data_sorted[:-k]
    elif tail == 'smallest':
        reduced_data = data_sorted[k:]
    else:
        raise ValueError("tail must be 'largest' or 'smallest'")
    
    mean_reduced = np.mean(reduced_data)
    sq_dev_reduced = np.sum((reduced_data - mean_reduced) ** 2)
    
    T_statistic = sq_dev_reduced / sq_dev_full
    critical_value = 1 - alpha  # Approximate
    
    if tail == 'largest':
        outliers = data_sorted[-k:]
    else:
        outliers = data_sorted[:k]
    
    return T_statistic, critical_value, outliers

# Set parameters
k = 3       # Number of suspected outliers
tail = 'largest'  # Or 'smallest'
alpha = 0.05

# Apply Tietjen-Moore Test to each column
for col in columns:
    print(f"\nColumn: {col}")
    data = df[col].dropna().values
    if len(data) <= k:
        print("Not enough data points to detect outliers.")
        continue
    T_stat, crit_val, outliers = tietjen_moore_test(data, k, tail, alpha)
    
    print(f"Tietjen-Moore Statistic: {T_stat:.4f}")
    print(f"Approximate Critical Value: {crit_val}")
    
    if T_stat < crit_val:
        print(f"Outliers detected ({k} suspected): {outliers}")
    else:
        print("No outliers detected.")

```
